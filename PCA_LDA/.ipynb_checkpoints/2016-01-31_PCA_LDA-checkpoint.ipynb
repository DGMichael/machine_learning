```python
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
#PCA:
iris = datasets.load_iris()
X = iris.data
y = iris.target
target_labels = iris.target_names
```

```python
pca = PCA(n_components=2)
X_reduced = pca.fit(X).transform(X)
```

```python
print "Explained variance by first two principal components: {0} and {1}".format(pca.explained_variance_ratio_[0],pca.explained_variance_ratio_[1])
```

```python
X_reduced[0:5]
```

```python
plt.figure()
plt.scatter(X_reduced[:,0], X_reduced[:,1], c=y)
plt.legend()
```

Assess output of KNN classifier using full and PCA reduced data:

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
from sklearn import cross_validation

neigh = KNeighborsClassifier(n_neighbors=3)
predicted_full = cross_validation.cross_val_predict(neigh,iris.data, iris.target, cv=10)
print "Accuracy with full feature vectors {0}".format(metrics.accuracy_score(iris.target,predicted_full))
```

```python
predicted_PCA = cross_validation.cross_val_predict(neigh,X_reduced, iris.target, cv=10)
print "Accuracy with full feature vectors {0}".format(metrics.accuracy_score(iris.target,predicted_PCA))
```

Apparently there is no improvement with selected parameters using PCA reduced data.

Running Linear Discriminant Analysis on the iris dataset:
-Used to identify features that account for the most variance between classes.  Might be better for use in identifying dimensions for classifier training.

```python
lda = LinearDiscriminantAnalysis(n_components=2)
X_reduced_LDA = lda.fit(X,y).transform(X)
```

```python
plt.scatter(X_reduced_LDA[:,0], X_reduced_LDA[:,1], c=y)
```

```python
predicted_LDA = cross_validation.cross_val_predict(neigh,X_reduced_LDA, iris.target, cv=10)
print "Accuracy with full feature vectors {0}".format(metrics.accuracy_score(iris.target,predicted_LDA))
```

LDA improved accuracy using 5 fold cross-validation by ~1 percent.
